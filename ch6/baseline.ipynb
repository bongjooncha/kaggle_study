```python
#데이터 불러오기

import numpy as np
import pandas as pd

data_path = './data/'
train = pd.read_csv(data_path + 'train.csv')
test = pd.read_csv(data_path + 'test.csv')
submission = pd.read_csv(data_path + 'sampleSubmission.csv')
```

```python
# weather이 4인경우는 이상치였기에 제거
train = train[train['weather']!=4]
```

```python
# train데이터와 test데이터 합치기
all_data = pd.concat([train, test], ignore_index=True) #ignore_index를 통해서 원래 index를 무시하고 새롭게 index를 생성(train과 test를 합쳤기에)
all_data
```

```python
all_data['datetime'] = pd.to_datetime(all_data['datetime'])
all_data['date'] = all_data['datetime'].dt.date
all_data['year'] = all_data['datetime'].dt.year
all_data['month'] = all_data['datetime'].dt.month
all_data['hour'] = all_data['datetime'].dt.hour
all_data['weekday'] = all_data['datetime'].dt.weekday
all_data
```

```python
# 삭제할 데이터
drop_features = ['datetime', 'casual', 'registered', 'date', 'month', 'windspeed']
```

```python
all_data = all_data.drop(columns=drop_features)
all_data

```

```python
# 데이터 나누기
X_train = all_data[~pd.isnull(all_data['count'])]
X_test = all_data[pd.isnull(all_data['count'])]

X_train = X_train.drop(['count'],axis=1)
X_test = X_test.drop(['count'],axis=1)

y = train['count']
# X_train.head()
X_test.head()
```

```python
import numpy as np

def rmsle(y_true, y_pred, convertExp=True):
    if convertExp:
        y_true = np.exp(y_true)
        y_pred = np.exp(y_pred)
    log_true = np.nan_to_num(np.log(y_true+1))
    log_pred = np.nan_to_num(np.log(y_pred+1))
    
    return np.sqrt(np.mean((log_true - log_pred)**2))

```

```python
from sklearn.linear_model import LinearRegression

linear_reg_model = LinearRegression()

log_y = np.log(y)
linear_reg_model.fit(X_train, log_y)

pred = linear_reg_model.predict(X_train)

print(f'선형 회구의 RSMSLE 값: {rmsle(log_y, pred, True):.4f}')
```

```python
# 예측결과 도출
linearreg_preds = linear_reg_model.predict(X_test)
submission['count'] = np.exp(linearreg_preds)
submission.to_csv('submission.csv', index=False)
```
