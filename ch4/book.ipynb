<h3>4.1.1 수치형 데이터</h3>

<b>수치형 데이터(numerical data)</b>는 사칙 연산이 가능한 데이터. data type으로는 int, float, bool 등이 있다. (bool은 0과 1로 표현되기에 사칙 연산이 가능.)


<li>연속형 데이터(continuous data): 데이터가 연속적으로 이어진다. 예를 들어, 키, 몸무게, 온도 등이 있다.</li>
<br>
<li>이산형 데이터(discrete data): 데이터가 이산적으로 이어진다. 예를 들어, 사람의 수, 소의 수, 차의 수 등이 있다.</li>
<br>

*연속형은 실수형, 이산형은 정수형 데이터*

<h3>4.1.2 범주형 데이터</h3>

<b>범주형 데이터(categorical data)</b>는 범주를 나눌 수 있는 데이터이며 사칙 연산이 불가능한 데이터. data type으로는 object, category 등이 있다.(여기서 object와 category는 pandas라이브러리에서 사용됨. object는 string type과 동일하지만, category는 고유한 값을 가진 데이터로 효율 적으로 메모리 관리 할 수 있께 해줌)


<li>순서형 데이터(ordinal data)</b>순위를 매길 수 있는 데이터. 예를 들어, 학점(A, B, C, D) 등이 있다.</li>
<br>
<li>명목형 데이터(nominal data)</b>순위가 따로 존재하지 않는 데이터. 예를 들어, 성별(남, 여) 등이 있다. 또한 우편번호 역시 숫자로 되어있지만 사칙연산과 관련 없고 순위가 존재하지 않기 때문에 명목형으로 분류된다.</li>
<br>



```python
import sys
import pandas as pd

obj_data1 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"])
cat_data1 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"], dtype="category")
obj_data2 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"])
cat_data2 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"],
                      dtype="category")

print(sys.getsizeof(obj_data1))  # object1 타입의 메모리 사용량
print(sys.getsizeof(cat_data1))  # category1 타입의 메모리 사용량
print(sys.getsizeof(obj_data2))  # object2 타입의 메모리 사용량
print(sys.getsizeof(cat_data2))  # category2 타입의 메모리 사용량
```

<h3>4.3.1 히스토그램</h3>

수치형 데이터의 구간별 빈도수를 나타내는 그래프. (historical의 histo-와 gram의 합성어. 역사적 데이터 분포를 데이터 형태로 시각적으로 표현)

```python
sns.histplot(data= titanic, x= 'age')
```

<li>bins는 구간을 정의(기본값은 10)</li>
</br>
<li>hue는 데이터의 추가적인 범주형을 지정. 다른 색성으로 이를 표현</li>
</br>
<li>multiple은 여러개의 변수를 누적해서 표현</li>

```python
sns.histplot(data=titanic, x= "age", bins=10)
```

```python
sns.histplot(data=titanic, x= "age", hue='alive')
# 회색구간은 alive no와 yes의 공통 부분
```

```python
sns.histplot(data=titanic, x= "age", hue='alive', multiple = 'stack')
```

<h3>4.3.2 커널밀도추정 함수 그래프(kdeplot)</h3>

데이터 분포를 나태내는 함수로 특정 구간 데이터가 존재할 확률을 나타냄


<img src="../ch4/img/kde.png" alt="kde">


```python
sns.kdeplot(data=titanic, x='age')
```

```python
sns.kdeplot(data=titanic, x='age',hue='alive', multiple='stack')
```

<h3>4.3.3 분포도(displot)</h3>

kaggle에서 자주 사용하는 분포도

hisplot이 그린 히스토그램과 크게 다르지 않음

kind에 'kde'를 사용할 시 커널밀도함수로 사용가능


```python
sns.displot(data=titanic, x='age', kind='kde')
```

```python
sns.displot(data=titanic, x='age', kde=True)
```

<h3>4.3.4 러그플롯(rugplot)</h3>

러그플롯은 주변 분포(marginal distribution)을 나태나는 그래프. 

```python
sns.rugplot(data=titanic, x = 'age')
```

```python
sns.kdeplot(data=titanic, x= 'age')
sns.rugplot(data=titanic, x = 'age')
```

```python
sns.rugplot(data=titanic, x = 'age')
sns.kdeplot(data=titanic, x= 'age')
```

두개의 함수가 겹쳐지는 경우 먼저 그려지는 함수에 의해서 배경이 결정.
